```python
combined=[a+b for a in "wow" for b in "life"]
print(combined)
#creates every possible substring of 2 letters
```

```python
#nested list comprehensions
nested=[letters[1] for letters in[a+b for a in "wow" for b in "life"]]
nested
```

```python
#dictionary comprehension
words=["life","wow","study"]
my_dict={word:len(word) for word in words}
my_dict
```

```python
#zip makes tuple pairs of values from 2 different lists
words=["life","wow","study"]
ind=[1,2,3]
pairs=zip(words,ind)
print(pairs)
for i in pairs:
    print(i)
```

```python
my_dict2={key:value for (key,value) in zip(words,ind)}
```

```python
my_dict2
```
